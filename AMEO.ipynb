## Regression - Base Model:

```python
data = df.drop(['Designation','JobCity','10board','12board','Specialization'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
data['Degree']=le.fit_transform(data['Degree'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
X = data.drop('Salary',axis=1)
y = data['Salary']
```

```python
import statsmodels.api as sm
X_const = sm.add_constant(X)
result = sm.OLS(y,X_const).fit()
result.summary()
```

```python
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
x_bfe =data[selected_features_BE]
y_bfe = data['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_bfe,y_bfe,test_size=0.30,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
print('R2_Score: ',r2_score(ytest,ypred))
print('RMSE: ',np.sqrt(mean_squared_error(ytest,ypred)))
print('MAPE: ',np.mean(np.abs((ytest - ypred) / ytest)) * 100)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor

lr = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
bag = BaggingRegressor()
ada = AdaBoostRegressor()
gb = GradientBoostingRegressor()
```

```python
X = data.drop('Salary',axis=1)
y = data['Salary']
```

```python
import statsmodels.api as sm
X_const = sm.add_constant(X)
result = sm.OLS(y,X_const).fit()
result.summary()
```

```python
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
x_bfe =data[selected_features_BE]
y_bfe = data['Salary']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_bfe,y_bfe,test_size=0.30,random_state=42)
```

```python
from sklearn.metrics import r2_score
models = [lr,dt,rf,bag,ada,gb]
def salary(a,b,c,d):
    for i in models:
        i.fit(a,c)
        ypred = i.predict(b)
        print('Model: {}'.format(i))
        print('R2_Score: ',r2_score(d,ypred))
        print('RMSE: ',np.sqrt(mean_squared_error(d,ypred)))
        print('MAPE: ',np.mean(np.abs((d - ypred) / d)) * 100)
```

```python
salary(xtrain,xtest,ytrain,ytest)
```
